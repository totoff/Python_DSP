```python
import numpy as np
import matplotlib.pyplot as plt
t1 = 1
t2 = 11
f1= 6000
f2 = 5000
DI = t2 - t1
deltaF = f2 - f1
a = -t1*deltaF - t1**2*deltaF/DI
b = t1*deltaF/DI + f2

t = np.linspace(t1,t2,100,endpoint=True)
mod = a/t + b
plt.plot(t,mod)
```

```python
import numpy as np
import matplotlib.pyplot as plt
t1 = 0
t2 = 4095
t = np.linspace(t1,t2,4096,endpoint=True)
a = -2920.4
b = -5236.74
c = -1.5576
mod = a/(t + b) + c
plt.plot(t,mod)
```

```python
f0 = 2000
deltaF = 1000
DI = 2000
nbEchantillons = DI/1000 * 48000
t = np.arange(0, nbEchantillons)
t = t/nbEchantillons * 4096
f = f0 + deltaF * (a/(t + b) + c)
treel = DI/4096*t 
plt.plot(treel,f)
```

```python
import IPython.display as ipd
sig=np.sin(2*np.pi*f*treel/1000)
ipd.Audio(sig,rate=48000)
```

```python
from scipy import signal
f1, t1, Sxx = signal.spectrogram(sig, fs=48000,  mode='psd')
plt.pcolormesh(t1, f1, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
f0 = 3000
deltaF = 2000
DI = 2000
nbEchantillons = DI/1000 * 48000
t = np.arange(0, nbEchantillons)
t = t/nbEchantillons * 4096
tprime = 4095 - t
f = f0 + deltaF * (a/(tprime + b) + c)
treel = DI/4096*t 
plt.plot(treel,f)
```

```python
import IPython.display as ipd
sig=np.sin(2*np.pi*f*treel/1000)
ipd.Audio(sig,rate=48000)
```

```python
from scipy import signal
f1, t1, Sxx = signal.spectrogram(sig, fs=48000, nperseg=256)
plt.pcolormesh(t1, f1, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
