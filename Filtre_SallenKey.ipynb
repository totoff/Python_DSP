```python
import numpy
from  matplotlib.pyplot import *

C=10e-9
R=22e3
m=numpy.sqrt(2)
K=3-m
fc=1.0/(1*numpy.pi*R*C)
def H(f):
    return K/(1+1j*m*f/fc-(f/fc)**2)
def bode(H,start,stop):
    freq = numpy.logspace(start=start,stop=stop,num=1000)
    h = H(freq)
    gdb = 20*numpy.log10(numpy.absolute(h))
    phi = numpy.angle(h)
    figure(figsize=(8,8))
    subplot(211)
    plot(freq,gdb)
    xscale('log')
    xlabel("f (Hz)")
    ylabel("GdB")
    grid()
    subplot(212)
    plot(freq,phi)
    xscale('log')
    xlabel("f (Hz)")
    ylabel("phi")
    grid()
bode(H,1,5)
```
